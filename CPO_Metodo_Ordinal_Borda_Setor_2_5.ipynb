---

visao_geral.png

Conforme descrito por Helder Homes (2014), a Enciclopdia Britnica (Encyclopaedia Britannica, 2012) registra que o mtodo de Borda foi apresentado por Jean-Charles Borda em 1781, na Frana para ser aplicado em comits compostos por mais de um indivduo (problema multidecisor). A ideia central deste mtodo consiste em estabelecer uma combinao das ordenaes ou *ranking* individuais estabelecidos por cada um dos decisores em um *ranking* global.

Por fim, cabe destacar que o Mtodo de Borda  do tipo **problemtica &gamma;**

## &#9776; **CONSTANTES E VARIVEIS GLOBAIS DO MDULO**
---

```python
# cria as variveis globais e constantes
AVALIACOES = ["E","MB","B","R","I"]
PONTUACOES_MAX_GEO_Q2 = {"E" : 1, "MB" : 2, "B" : 4, "R" : 8, "I" : 16}
```

## &#9776; **MTODOS DO MDULO**
---

## **&#x2609; listarCriterios()**

Funo para listar os critrios e quantidades existentes.

```python
def listarCriterios(df):
  qtd_criterios = df.shape[1]
  criterios = df.columns.tolist() 
  return criterios, qtd_criterios
```

## **&#x2609; listarAlternativas()**

Funo para listar as alternativas e quantidades existentes.

```python
def listarAlternativas(df):
  qtd_alternativas = df.shape[0]
  alternativas = df.index.tolist()
  return alternativas, qtd_alternativas
```

## **&#x2609; avaliarAlternativas()**

```python
def avaliarAlternativas(df):
  
  criterios, qtd_criterios = listarCriterios(df)
  alternativas, qtd_alternativas = listarAlternativas(df)
  print(f"\nOBSERVAO:\n Antes de inicar sua avaliao, leia atentamente o significado de cada critrio.\n")
  mostrarDicionario()
  print("\nPara continuar aperte [ENTER]\n")
  espera = input()
  # apresenta alternativas e critrios a serem avaliados
  # decisor pontua com E, MB, B, R, I
  for criterio in criterios:
    for alternativa in alternativas:
      avaliacao = input(f"CRITRIO [{criterio}] ALTERNATIVA [{alternativa}]\nAVALIAES PERMITIDAS : E, MB, B, R, I\nSua avaliao: ") 
      while avaliacao.isnumeric() or avaliacao not in AVALIACOES:
        print(f"\nDigitar apenas E, MB, B, R ou I!\n")
        avaliacao = input(f"CRITRIO [{criterio}] ALTERNATIVA [{alternativa}]\nAVALIAES PERMITIDAS : E, MB, B, R, I\nSua avaliao: ")
      # lana avaliao 
      df.loc[alternativa,criterio]= avaliacao
    print("")
  return df
```

## **&#x2609; obterChave()**

Altera avaliao qualitativa para um sistema de pontuao quantitativo.

```python
def obterChave(avaliacao, dicionario = PONTUACOES_MAX_GEO_Q2):
  return dicionario.get(avaliacao)
```

## **&#x2609; transformarAvaliacoes()**

```python
def transformarAvaliacoes(df):
  # obtem critrios e alternativas 
  criterios, qtd_criterios = listarCriterios(df)
  alternativas, qtd_alternativas = listarAlternativas(df)

  for criterio in criterios:
    for alternativa in alternativas:
      pontuacao = obterChave(df.loc[alternativa,criterio])
      df.loc[alternativa,criterio] = pontuacao  
  return df
```

## **&#x2609; somarPontos()**

Funo que armazena no atributo [total_pontos] os pontos referentes de cada alternativa. 

```python
def somarPontos(df):
  df['total_pontos'] = df.sum(axis=1)
  return df
```

## **&#x2609; salvarResultadoFinalDecisor()**

```python
def salvarResultadoFinalDecisor(projeto,decisor,df):
  os.chdir("/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/" + projeto + "/resultados")
  try:
    df.to_csv("decisor" + decisor + ".csv", sep=",", encoding="utf-8")
    msg = "Resultado armazenado com sucesso!"
  except:
    msg = "Problemas ao armazenar resultado!"
  return msg
```

## **&#x2609; criarProjeto()**

Funo utilizada no Mdulo Administrador para criar o Projeto.

```python
def criarProjeto(projeto):
  os.chdir("/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda")
  try:
    if not os.path.exists(projeto):
      os.mkdir(projeto)
      os.mkdir(projeto + "/avaliacoes")
      os.mkdir(projeto + "/decisores")
      os.mkdir(projeto + "/resultados")
      msg = "Projeto criado com sucesso!"
    else :
      msg = "Projeto j existe!"
  except:
    msg = "Problemas ao criar projeto!"
  return msg
```

## **&#x2609; gerarArquivosDecisores()**

Funo utilizada no Mdulo Administrador para gerar os arquivos que sero trabalhados por cada Decisor.

```python
def gerarArquivosDecisores(projeto, qtd_decisores, qtd_projetos):
  os.chdir("/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/" + projeto + "/decisores")
  projetos = []
  for p in range(1, qtd_projetos+1):
    projetos.append("P" + str(p))
  df = pd.DataFrame(data = {"projetos" : projetos}, columns = ["projetos", "c1","c2","c3","c4","c5"])
  df.set_index(df.columns[0], inplace= True, drop=True)
  decisores = [str(x) for x in range(1,qtd_decisores+1)]                    
  for decisor in decisores:
    try:
      df.to_csv("decisor" + decisor + ".csv")
      msg = "Arquivos gerados com sucesso para os Decisores"
    except:
      msg = "Problemas ao gerar arquivos para os Decisores!"
  return msg
```

## **&#x2609; contarArquivosResultados**

Funo utilizada no Mdulo Administrador para apresentar quantidade de decisores para o projeto criado.


```python
def contarArquivosResultados(projeto):
  URL = "/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/"+ projeto + "/resultados"
  os.chdir(URL)
  contador = 0
  for path in pathlib.Path(".").iterdir():
    if path.is_file():
        contador += 1
  return str(contador)
```

## **&#x2609; concatenarResultados**

Funo utilizada no Mdulo Administrador para apresentar quantidade de decisores para o projeto criado.


```python
def concatenarResultados(projeto):
  URL = "/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/"+ projeto + "/resultados"
  os.chdir(URL)
  filepaths = [f for f in os.listdir(".") if f.endswith('.csv')]
  df = pd.concat(map(pd.read_csv, filepaths), axis=0, ignore_index=True)
  return df
```

## **&#x2609; carregarAdministrarProjeto()**

Funo para inicializar o Mdulo de Administrao.

```python
def carregarAdministrarProjeto():
  # criar projeto
  carregarCabecalhoModuloAdminbistrador("MDULO ADMINISTRADOR - ADMINISTRAR PROJETOS")
  # recebe nome do projeto para criar pasta
  projeto = input("\nInforme o nome do Projeto: ")
  print(f"{criarProjeto(projeto)}")
  # recebe nmero de decisores para gerar arquivos
  qtd_decisores = input("\nInforme o nmero de decisores: ")
  while qtd_decisores.isalpha() :
      qtd_decisores = input("Informe nmero de decisores: ")
  # recebe o nmero de projetos para defnir dataframe decisores
  qtd_projetos = input("\nInforme o nmero de projetos: ")
  # aguarda resposta da execuo das tarefas 
  print(gerarArquivosDecisores(projeto, int(qtd_decisores), int(qtd_projetos)))
```

## **&#x2609; mostrarResultadoIndividual()**

```python
def mostrarResultadoIndividual(projeto, qtd_decisores):
  os.chdir("/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/"+ projeto + "/resultados")
  for decisor in range(1,int(qtd_decisores)+1):
    arquivo = "decisor" + str(decisor) + ".csv"
    df = pd.read_csv(arquivo)
    print(f"Avaliao Decisor {decisor}")
    display(df)
```

## **&#x2609; mostrarResultadoFinal()**

```python
def mostrarResultadoFinal(projeto):
  df_resultados = concatenarResultados(projeto)
  df_resultados = df_resultados.groupby("projetos", sort=True)["total_pontos"].sum()
  display(df_resultados)
```
