## &#9776; **MTODO ORDINAL MULTICRITRIO DE BORDA**
---

visao_geral.png

Conforme descrito por Helder Homes (2014), a Enciclopdia Britnica (Encyclopaedia Britannica, 2012) registra que o mtodo de Borda foi apresentado por Jean-Charles Borda em 1781, na Frana para ser aplicado em comits compostos por mais de um indivduo (problema multidecisor). A ideia central deste mtodo consiste em estabelecer uma combinao das ordenaes ou *ranking* individuais estabelecidos por cada um dos decisores em um *ranking* global.

Por fim, cabe destacar que o Mtodo de Borda  do tipo **problemtica &gamma;**

## &#9776; **ETAPAS PARA APLICAO DO MTODO ORDINAL MULTICRITRIO DE BORDA**
---

Os seguintes passos devero ser efetuados na aplicao do mtodo de borda:

**Passo 1** : Ordenar as alternativas

**Passo 2** : A cada posio da alternativa  atribuda uma pontuao correspondente (1 lugar = 1 ponto; 2 lugar = 2 pontos; e assim sucessivamente).

**Passo 3** : Somar todos os pontos ganhos para casa alternativa

**Passo 4** : As alternativas so ordenadas de forma crescente, sendo a melhor alternativa aquela que obteve a menor pontuao.

## **&#x2609; cabecalhoSMB()**

Funo para imprimir cabealho

```python
def cabecalhoMB():
  print("="*90)
  print("SISTEMA PARA SELEO DE ALTERNATIVAS BASEADO NO MTODO DE BORDA".center(90))
  print("="*90 + "\n")
```

## **&#x2609; carregarDataFrame()**

Funo utilizada no mdulo Decisor.

```python
def carregarDataFrame(arquivo, separador=','):
  df = pd.read_csv(arquivo + ".csv", sep=separador)
  df.set_index(df.columns[0], inplace= True)
  return df
```

## **&#x2609; mostrarDataFrame()**

Funo para carregar mltiplos DataFrames.

```python
def mostrarDataFrame(df):
  display(df.head())
```

## **&#x2609; listarCriterios()**

Funo para listar os critrios e quantidades existentes.

```python
def listarCriterios(df):
  qtd_criterios = df.shape[1]
  criterios = df.columns.tolist() 

  return criterios, qtd_criterios
```

## **&#x2609; listarAlternativas()**

Funo para listar as alternativas e quantidades existentes.

```python
def listarAlternativas(df):
  qtd_alternativas = df.shape[1]
  alternativas = df.index.tolist()

  return alternativas, qtd_alternativas
```

## **&#x2609; ranquearAlternativas()**

Funo para obter as avaliaes ou julgamentos estabelecidos por cada decisor para cada uma das alternativas.

```python
def ranquearAlternativas(df):
  
  criterios, qtd_criterios = listarCriterios(df)
  alternativas, qtd_alternativas = listarAlternativas(df)
  intervalo_pontos = [str(x) for x in range(1,qtd_alternativas+1)]
  
  cabecalhoMB()
  print(f"MATRIZ DE DECISO A SER PONTUADA".center(90))
  mostrarDataFrame(df_cenario)
  print("")
  for criterio in criterios:
    for alternativa in alternativas:
      pontuacao = input(f"CRITRIO [{criterio}] ALTERNATIVA [{alternativa}]: ")
      while pontuacao.isalpha() or pontuacao not in intervalo_pontos:
        print(f"\nDigitar apenas nmeros ou ranqueamento permitido!\n")
        print(f"\nRanqueamento ainda permitido para este critrio {intervalo_pontos}\n")
        pontuacao = input(f"CRITRIO [{criterio}] ALTERNATIVA [{alternativa}]: ")
      df.at[alternativa,criterio] = int(pontuacao)
      intervalo_pontos.remove(pontuacao)
    intervalo_pontos = [str(x) for x in range(1,qtd_alternativas+1)]
    print("")

  print(f"MATRIZ DE DECISO RANQUEADA".center(90))
  display(df.head())

  return df
```

## **&#x2609; somarPontos()**

Funo que armazena no atributo [total_pontos] os pontos referentes de cada alternativa. 

```python
def somarPontos(df):
  df['total_pontos'] = df.sum(axis=1)
  return df
```

## **&#x2609; apresentarDecisorResultado()**

Funo para apresentar resultado ao Decisor.

```python
def apresentarDecisorResultado(df):
  selecao_alternativas = df.sort_values(by='total_pontos',ascending=True)
  prioridades = selecao_alternativas.index
  cabecalhoMB()
  print("Alternativa para Aquisio\n")
  print(f"Alternativa selecionada pelo Decisor : {prioridades[0]}")
  return selecao_alternativas.iloc[0,0]
```

## **&#x2609; guardarDecisorResultado()**

Funo para salvar resultado final do Decisor.

```python
def guardarDecisorResultado(alternativa, projeto, decisor):
  df_resultado = pd.DataFrame({'DECISOR' : [],
                               'ALTERNATIVA' : []})
  df_resultado.loc[0,"decisor"] = "Decisor" + decisor
  df_resultado.loc[0,"alternativa"] = str(alternativa)
  arquivo = "Decisor" + decisor + ".csv"
  try:
    df_resultado.to_csv( arquivo, sep=",", encoding="utf-8")
    msg = "Resultado armazenado com sucesso!"
  except:
    msg = "Problemas ao armazenar resultado!"
  return msg
```

## **&#x2609; criarProjeto()**

Funo utilizada no Mdulo Administrador para criar o Projeto.

```python
def criarProjeto(projeto):
  try:
    if not os.path.exists(projeto):
      os.mkdir(projeto)
      os.mkdir(projeto + "/decisores")
      os.mkdir(projeto + "/resultado")
      msg = "Projeto criado com sucesso!"
    else :
      msg = "Projeto j existe!"
  except:
    msg = "Problemas ao criar projeto!"
  return msg
```

## **&#x2609; gerarArquivosDecisores()**

Funo utilizada no Mdulo Administrador para gerar os arquivos que sero trabalhados por cada Decisor.

```python
def gerarArquivosDecisores(projeto, qtd_decisores, df):
  URL = "/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/" + projeto + "/decisores/"
  os.chdir(URL)
  decisores = [str(x) for x in range(1,qtd_decisores+1)]                    
  for decisor in decisores:
    try:
      df.to_csv("Decisor" + decisor + ".csv")
      msg = "Arquivos gerados com sucesso para os Decisores"
    except:
      msg = "Problemas ao gerar arquivos para os Decisores!"
  return msg
```

## **&#x2609; contarArquivosResultados**

Funo utilizada no Mdulo Administrador para apresentar quantidade de decisores para o projeto criado.


```python
def contarArquivosResultados(projeto):
  URL = "/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/"+ projeto + "/resultado"
  os.chdir(URL)
  contador = 0
  for path in pathlib.Path(".").iterdir():
    if path.is_file():
        contador += 1
  return str(contador)
```

## **&#x2609; contarArquivosDecisores**

Funo utilizada no Mdulo Administrador para apresentar quantidade de decisores que j escolheram sua alternativa.

```python
def contarArquivosDecisores(projeto):
  URL = "/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/"+ projeto + "/decisores"
  os.chdir(URL)
  contador = 0
  for path in pathlib.Path(".").iterdir():
    if path.is_file():
        contador += 1
  return str(contador)
```

## **&#x2609; mostrarAdmVotosDecisores()**

Funo que apresenta um DataFrame nico com todos os votos j realizados.

```python
def mostrarAdmVotosDecisores(projeto):
  URL = "/content/drive/MyDrive/Pesquisa_Operacional/MetodoBorda/"+ projeto + "/resultado"
  os.chdir(URL)
  filepaths = [f for f in os.listdir(".") if f.endswith('.csv')]
  df = pd.concat(map(pd.read_csv, filepaths))
  return df[["decisor","alternativa"]]
```

## **&#x2609; carregarAdministrarProjeto()**

Funo para inicializar o Mdulo de Administrao.

```python
def carregarAdministrarProjeto(df):
  # criar projeto
  print("="*90)
  print("MDULO ADMINISTRADOR".center(90))
  print("="*90)
  projeto = input("\nInforme o nome do Projeto: ")
  print(f"{criarProjeto(projeto)}")

  # define nmero de decisores
  qtd_decisores = input("Informe nmero de decisores: ")
  while qtd_decisores.isalpha() :
      qtd_decisores = input("Informe nmero de decisores: ")
  # gera tokens para os decisores
  print(gerarArquivosDecisores(projeto, int(qtd_decisores), df))
```

## **&#x2609; carregarAdministrarResultado()**

Funo que mostra Dashboard com o resultado final e melhor alternativa para o projeto.

```python
from matplotlib.pyplot import title
def carregarAdmMostrarResultado(projeto):
  qtd_decisores = contarArquivosDecisores(projeto)
  qtd_decisores_comvotos = contarArquivosResultados(projeto) 
  df_votos_ind = mostrarAdmVotosDecisores(projeto)
  alternativa_final =  pd.value_counts(df_votos_ind['alternativa'].values, sort=True)
  display(HTML('<h1>:: Dashboard - Sistema Mtodo de Borda</h1><hr/>'))
  display(HTML('<table width="100%"><tr><td width="33%"><center><h2 style="color:#1EA8E7"><b>Projeto</b></h2></center></td><td width="33%"><center><h2 style="color:#1EA8E7"><b>Nmero de Decisores</b></h2></center></td><td width="33%"><center><h2 style="color:#1EA8E7"><b>Nmero de Votos</b></h2></center></td></tr></table> '))
  display(HTML('<table width="100%"><tr ><td width="33%"><center><h2><b>' + projeto + '</b></h2></center></td><td width="33%"><center><h2 ><b>' + qtd_decisores + '</b></h2></center></td><td width="33%"><center><h2><b>' + qtd_decisores_comvotos + '</b></h2></center></td></tr></table> '))
  display(HTML('<center><img src="https://i.postimg.cc/MK44rDCY/admin.png" /><br/><p><b>ADMIN</b></p></center>'))
  display(HTML('<h2 style="color:#1EA8E7">Resultado Individual por Decisores</h2>'))
  display(df_votos_ind)
  display(HTML('<hr/>'))
  display(HTML('<h2 style="color:#1EA8E7">Representao Grfica dos Votos</h2>'))
  display(alternativa_final.plot.bar(title="RESULTADO FINAL", rot=0, color = ["blue"]))
```
